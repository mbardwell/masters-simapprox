## import data

```python
buses = pd.read_csv(path_to_scigrid_data + "buses.csv")
loads = pd.read_csv(path_to_scigrid_data + "loads-p_set.csv")
vmags = pd.read_csv(path_to_scigrid_data + "vmags.csv")
vangs = pd.read_csv(path_to_scigrid_data + "vangs.csv")
qmags = pd.read_csv(path_to_scigrid_data + "qmags.csv")
linemags = pd.read_csv(path_to_scigrid_data + "lines-p0-mags.csv")

def personalise_column_names(df, name):
    new_columns = []
    for column in df.columns:
        new_columns.append(name +  "-" + str(column))
    df.columns = new_columns

personalise_column_names(loads, "load")
personalise_column_names(vmags, "vmag")
personalise_column_names(vangs, "vang")
personalise_column_names(vangs, "qmags")
personalise_column_names(vangs, "linemags")

features = loads.drop("load-name", axis=1)
labels = vmags.drop("vmag-name", axis=1)
full = features.join(labels)

n_samples = features.shape[0]
n_features = features.shape[1]
n_labels = labels.shape[1]

print("n_features: {}, n_labels: {}, n_samples: {}".format(
    n_features,
    n_labels,
    n_samples))
```

## feature exploration

```python
# bus-to-bus l2 distance

l2_df = pd.DataFrame(columns=buses.name)
for i, bus in enumerate(buses.ix[:,0]):
    ref_point = buses[["x", "y"]].loc[i,:]
    l2_df[bus] = buses[["x", "y"]].sub(np.array(ref_point)).pow(2).sum(1).pow(0.5)
l2_df.insert(loc=0, column="bus", value=pd.Series(buses.name, index=l2_df.index))
display(l2_df)
```

```python
# many of the buses have vmag = 1.0 (PV-buses) for each sample. Remove these

if "vmag-1" in labels.keys():
    columns_to_remove = []
    for name in vmags.columns:
        remove = True
        for value in vmags[name]:
            if value != 1.0:
                remove = False
        if remove:
            columns_to_remove.append(name)
        
    full = full.drop(columns_to_remove, axis=1)
    labels = labels.drop(columns_to_remove, axis=1)
    n_labels = n_labels-len(columns_to_remove)

    print("n_features: {}, n_labels: {}, n_samples: {}".format(
        n_features,
        n_labels,
        n_samples))
```

```python
# feature/target correlation

corr_matrix = full.corr()
cols = [c for i, c in enumerate(corr_matrix.columns) if corr_matrix.keys().str.contains("^load", regex=True)[i]]
reduced_corr_matrix = corr_matrix[cols]
reduced_corr_matrix = reduced_corr_matrix[~reduced_corr_matrix["load-1"].keys().str.contains("load")]

display(reduced_corr_matrix)
# pd.set_option('display.max_rows', 500)
# reduced_corr_matrix["load-1"].sort_values()
```

## preprocessing

```python
# scale features
scaled_features = pd.DataFrame(MinMaxScaler().fit_transform(features), columns=list(features.columns.values))
features = scaled_features
# scaled_labels = pd.DataFrame(MinMaxScaler().fit_transform(labels), columns=list(labels.columns.values))
# labels = scaled_labels
```

## training

```python
training_percentage = 80
n_training_samples = int(n_samples*(training_percentage/100))

random_seed=0
X_train = features.sample(n_training_samples, random_state=random_seed)
y_train = labels.sample(n_training_samples, random_state=random_seed)a
X_test = features[~features.isin(X_train)].dropna()
y_test = labels[~labels.isin(y_train)].dropna()

X_train = X_train.values
y_train = y_train.values
X_test = X_test.values
y_test = y_test.values
```

```python
training_stats = {"n_training_samples": n_training_samples}
stats = {"mae": [],
         "score": [], 
         "scorevar": [],
         "valscore": [],
         "time": []}
approx_type = {"rf": copy.deepcopy(stats), 
               "svr": copy.deepcopy(stats)}


# setup approximators
time_start = time.time()
forest = RandomForestRegressor()
forest.fit(X_train, y_train)
time_forest = time.time()

if n_training_samples > n_features+2:
    interp = LinearNDInterpolator(X_train, y_train.T[0].T)
time_interp = time.time()

svr = copy.deepcopy(stats)
for idx in range(n_labels):
    clf = SVR(gamma='scale', C=1.0, epsilon=0.2, kernel='rbf')
    clf.fit(X_train, y_train.T[idx].T)
    '''
    Scikit-Learn cross-validation features expect a utility function (greater is better) rather than a cost function
    (lower is better), so the scoring function is actually the opposite of the MSE (i.e., a negative value), 
    which is why the preceding code computes -scores before calculating the square root.
    - A. Geron, Hands on Machine Learning pg 101 
    '''
    svr_xval_score = cross_val_score(clf, X_train, y_train.T[idx].T,
                                     scoring="neg_mean_squared_error", cv=5, n_jobs=-1)
    svr_xval_score = np.sqrt(-svr_xval_score)
    svr["mae"].append(clf.predict(X_test)-y_test.T[idx])
    svr["score"].append(svr_xval_score.mean())
    svr["scorevar"].append(svr_xval_score.std())
    svr["valscore"].append(r2_score(clf.predict(X_test), y_test.T[idx]))
time_svr = time.time()

# score approximators
forest_xval_score = cross_val_score(forest, X_train, y_train, 
                                    scoring="neg_mean_squared_error", cv=5, n_jobs=-1)
forest_xval_score = np.sqrt(-forest_xval_score)
approx_type["rf"]["mae"].append(np.mean(forest.predict(X_test)-y_test))
approx_type["rf"]["score"].append(forest_xval_score.mean())
approx_type["rf"]["scorevar"].append(forest_xval_score.std())
approx_type["rf"]["valscore"].append(r2_score(forest.predict(X_test), y_test))
approx_type["rf"]["time"].append(time_forest-time_start)

if n_training_samples > n_features+2:
    approx_type["interp"] = copy.deepcopy(stats)
    interp_y_hat = interp(X_test)
    if n_labels == 1:
        not_nan_idx = [i for i, val in enumerate(interp_y_hat) if not np.isnan(val)]
    else:
        not_nan_idx = [i for i, val in enumerate(interp_y_hat) if not np.isnan(val[0])]
    interp_score = r2_score(y_test[not_nan_idx], interp_y_hat[not_nan_idx])
    approx_type["interp"]["mae"].append(np.mean(interp_y_hat[not_nan_idx] - y_test[not_nan_idx]))
    approx_type["interp"]["score"].append(interp_score)
    approx_type["interp"]["scorevar"].append(np.nan)
    approx_type["interp"]["time"].append(time_interp-time_forest)

approx_type["svr"]["mae"].append(np.mean(svr["mae"]))
approx_type["svr"]["score"].append(np.mean(svr["score"]))
approx_type["svr"]["scorevar"].append(svr_xval_score.std())
approx_type["svr"]["valscore"].append(np.mean(svr["valscore"]))
approx_type["svr"]["time"].append(time_svr-time_interp)

print("Training Stats: {}".format(training_stats))
for t in approx_type:
    print("\n", t + " Stats: \n{}".format(pd.DataFrame(approx_type[t])))
print("\n\n", "svr:\n", svr["score"])
```
